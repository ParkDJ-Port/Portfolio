```python
import gc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_listings = pd.read_csv('listings.csv', low_memory=False)
```

```python
aspect_listings = pd.read_csv('aspect_listings.csv')
```

```python
raw_listings['id'][]
```

```python
aspect_listings
```

```python
item_listings = raw_listings[['host_id',
                              'host_is_superhost',
                              'host_listings_count', 
                              'host_total_listings_count', 
                              'number_of_reviews', 
                              'number_of_reviews_ltm', 
                              'review_scores_rating',
                              'review_scores_accuracy',
                              'review_scores_cleanliness',
                              'review_scores_checkin',
                              'review_scores_communication',
                              'review_scores_location',
                              'review_scores_value',
                              'review_scores_value',
                              'property_type', 
                              'room_type']]
```

```python
g_1 = pd.merge(aspect_listings, item_listings, how='outer', on='host_id')
g_1.head()
```

```python
g_1.groupby('topic_5')['property_type'].value_counts()[0].index
```

```python
len(g_1['property_type'].value_counts())
```

```python
ax = sns.catplot(x='room_type', y='host_listings_count', hue='topic_5', kind='bar', data=g_1)
ax.fig.set_size_inches(15,8)
```

```python
# prop of property_type on each Aspect
data = g_1.groupby('topic_5')['property_type'].value_counts()

f, ax = plt.subplots(5, 1, figsize=(15,50))
for i in range(5):
    ax[i].pie(data[i].values, labels= data[i].index, autopct='%1.2f%%', shadow=True)
    ax[i].title.set_text(f'Aspect_{i}')
    
f.suptitle('Property_type')
plt.show()
```

```python
data = g_1.groupby('room_type')['topic_5'].value_counts()
```

```python
data
```

```python
data = g_1.groupby('room_type')['topic_5'].value_counts()

f, ax = plt.subplots(2,2, figsize=(15,15))
room_li = ['Entire home/apt', 'Hotel room', 'Private room', 'Shared room']
n=0

for i in range(2):
    for j in range(2):
        room_type = room_li[n]
        ax[i,j].pie(data[room_type].sort_index().values, labels=data[room_type].sort_index().index, autopct='%1.2f%%', shadow=True)
        n += 1
```

```python
# prop of room_type on each Aspect
data = g_1.groupby('topic_5')['room_type'].value_counts()
n=0

f, ax = plt.subplots(3, 2, figsize=(20,40))
for i in range(3):
    for j in range(2):
        if n==5:
            break
        else:
            ax[i,j].pie(data[n].values, labels= data[n].index, autopct='%1.2f%%', shadow=True)
            ax[i,j].title.set_text(f'Aspect_{n}')
        n+=1
        
f.suptitle('Room_type')
plt.show()
```

```python
data = g_1.groupby('topic_5')['property_type'].value_counts()
```

```python
# prop of property_type on each Aspect
data = g_1.groupby('topic_5')['property_type'].value_counts()
n=0

f, ax = plt.subplots(3, 2, figsize=(20,40))
for i in range(3):
    for j in range(2):
        if n==5:
            break
        else:
            ax[i,j].pie(data[n].values, labels= data[n].index, autopct='%1.2f%%', shadow=True)
            ax[i,j].title.set_text(f'Aspect_{n}')
        n+=1
        
f.suptitle('Property_type')
plt.show()
```

```python
# super host = 'false', proposition of aspects
f, ax = plt.subplots(1, 2, figsize=(15,15))

ax[0].pie(g_1.groupby('host_is_superhost')['topic_5'].value_counts()['f'].sort_index(), 
                labels=[0,1,2,3,4], autopct='%1.2f%%', shadow=True)
ax[1].pie(g_1.groupby('host_is_superhost')['topic_5'].value_counts()['t'].sort_index(), 
                labels=[0,1,2,3,4], autopct='%1.2f%%', shadow=True)
ax[0].title.set_text('Superhost=False')
ax[1].title.set_text('Superhost=True')
plt.show()
```

```python
single_listings = g_1[g_1['host_listings_count']==1]
```

```python
data = single_listings.groupby('room_type')['topic_5'].value_counts()
f, ax = plt.subplots(2,2, figsize=(15,15))
room_li = ['Entire home/apt', 'Hotel room', 'Private room', 'Shared room']
n=0

for i in range(2):
    for j in range(2):
        room_type = room_li[n]
        ax[i,j].pie(data[room_type].sort_index().values, labels=data[room_type].sort_index().index, autopct='%1.2f%%', shadow=True)
        ax[i,j].title.set_text(room_type)
        n += 1
```

```python
# prop of property_type on each Aspect
data = single_listings.groupby('topic_5')['property_type'].value_counts()
n=0

f, ax = plt.subplots(3, 2, figsize=(20,40))
for i in range(3):
    for j in range(2):
        if n==5:
            break
        else:
            ax[i,j].pie(data[n].values, labels= data[n].index, autopct='%1.2f%%', shadow=True)
            ax[i,j].title.set_text(f'Aspect_{n}')
        n+=1
        
f.suptitle('Property_type')
plt.show()
```

```python
data = single_listings.groupby('property_type')['topic_5'].value_counts()
property_li = single_listings['property_type'].value_counts().index
len_ = len(property_li)

f, ax = plt.subplots(int(len_/2), 2, figsize=(20,140))
n=0

for i in range(int(len_/2)):
    for j in range(2):
        property_type = property_li[n]
        ax[i,j].pie(data[property_type].sort_index().values, labels=data[property_type].sort_index().index, autopct='%1.2f%%', shadow=True)
        ax[i,j].title.set_text(property_type)
        n += 1
```
