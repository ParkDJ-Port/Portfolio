```python
r_li = pd.read_csv('listings.csv', low_memory=False)[['host_id',
                                                      'host_since', 
                                                      'host_response_time', 
                                                      'host_response_rate', 
                                                      'host_acceptance_rate', 
                                                      'host_is_superhost', 
                                                      'host_listings_count', 
                                                      'host_total_listings_count', 
                                                      'number_of_reviews', 
                                                      'number_of_reviews_ltm', 
#                                                      'review_scores_rating',
#                                                      'review_scores_accuracy',
#                                                      'review_scores_cleanliness',
#                                                      'review_scores_checkin',
#                                                      'review_scores_communication',
#                                                      'review_scores_location',
#                                                      'review_scores_value',
                                                      'price', 
                                                      'cancellation_policy', 
                                                      'property_type', 
                                                      'room_type']]
```

```python
r_li.isnull().sum()
```

```python
### 가장 최신 날짜
# r_li.sort_values(by='host_since', ascending = False)
```

```python
recent_day = '2020-05-05'
r_li['host_since'] = r_li['host_since'].apply(lambda x : (pd.Timestamp(recent_day) - pd.Timestamp(x)).days if x != 'NaN' else None)
```

```python
### 호스트 응답시간
```

```python
def response_time(line):
    if line == 'within an hour':
        line = 1
    elif line == 'within a few hours':
        line = 2
    elif line == 'within a day':
        line = 3
    elif line == 'a few days or more':
        line = 4
    else:
        line = 4
        
    return line
```

```python
r_li['host_response_time'] = r_li['host_response_time'].apply(lambda x : response_time(x))
```

```python
### 호스트 응답률, 성사율
```

```python
def percentage(line):
    line = re.sub('[%]', '', str(line))
    
    if line == 'nan':
        line = None
    else:
        line = int(line)*0.01
        
    return line
```

```python
r_li['host_response_rate'] = r_li['host_response_rate'].apply(lambda x : percentage(x))
r_li['host_acceptance_rate'] = r_li['host_acceptance_rate'].apply(lambda x : percentage(x))
```

```python
## 가격
```

```python
def price_type(line):
    line = re.sub('[$,]', '', str(line))
    line = float(line)
    return line
```

```python
r_li['price']
```

```python
r_li['price'] = r_li['price'].apply(lambda x : price_type(x))
```

```python
### 슈퍼호스트인지, 아닌지
```

```python
r_li['host_is_superhost'] = r_li['host_is_superhost'].apply(lambda x : 1 if x == 't' else (0 if x == 'f' else None))
```

```python
### 취소 정책
```

```python
r_li['cancellation_policy'] = r_li['cancellation_policy'].apply(lambda x : 0 if x == 'flexible' else (1 if x == 'moderate' else 2))
```

```python
tmp = []

for h, s in zip(g['host_id'], g['sentence_id']):
    tmp.append(str(h) + '_' + str(s))
    
g['h_s'] = tmp
```

```python
tmp4 = pd.DataFrame(g.groupby('host_id')['len_raw'].sum()).reset_index().rename(columns={'len_raw' : 'n_words_raw'})
```

```python
g2_tmp = g2.copy()
n_s = g2_tmp['n_sent'].to_numpy()
t_li = ['location', 'birth', 'introduction', 'attraction', 'interaction']

for t in t_li:
    g2_tmp[t+'_ps'] = g2_tmp[t].to_numpy()/n_s
```

## topic_ratio 추가 분석
- 비율(topic_ratio) = topic_n(각 토픽) / n_sent(문장의 갯수) : 각 토픽을 문장의 갯수로 나눠서
- room_type별로 topic_ratio의 평균을 구해서 그래프를 생성함

```python
g3 = g2.copy()
```

```python
g3 = g3.assign(location_ratio = g3['location']/g3['n_sent'])
g3 = g3.assign(birth_ratio = g3['birth']/g3['n_sent'])
g3 = g3.assign(introduction_ratio = g3['introduction']/g3['n_sent'])
g3 = g3.assign(attraction_ratio = g3['attraction']/g3['n_sent'])
g3 = g3.assign(interaction_ratio = g3['interaction']/g3['n_sent'])
```

```python
g3.head()
```

```python
df1 = pd.DataFrame(
    {'location_mean' : g3.groupby(['room_type'])['location_ratio'].mean(),
     'birth_mean' : g3.groupby(['room_type'])['birth_ratio'].mean(),
     'introduction_mean' : g3.groupby(['room_type'])['introduction_ratio'].mean(),
     'attraction_mean' : g3.groupby(['room_type'])['attraction_ratio'].mean(),
     'interaction_mean' : g3.groupby(['room_type'])['interaction_ratio'].mean(),
    }
)

df1.head()
```

```python
ax = df1.plot(kind = "bar")
ax.legend(bbox_to_anchor = (1.32, 1.005))

plt.show()
```
