```python
import gc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_listings = pd.read_csv('../datasets/listings.csv', low_memory=False)
```

```python
aspect_listings = pd.read_csv('../output/newyork/aspect_listings.csv')
```

```python
raw_listings.columns.to_list()
```

```python
item_listings = raw_listings[['host_id',
                              'host_is_superhost',
                              'host_listings_count', 
                              'host_total_listings_count', 
                              'number_of_reviews', 
                              'number_of_reviews_ltm', 
                              'review_scores_rating']]
```

```python
item_listings[item_listings['host_is_superhost']=='t']
```

```python
g_1 = pd.merge(aspect_listings, item_listings, on='host_id')
```

```python
g_1.groupby('host_is_superhost')['topic_5'].value_counts()
```

```python
g_1.groupby('host_is_superhost')['topic_5'].value_counts()['t'].sort_index()
```

```python
# super host = 'false', proposition of aspects
f, ax = plt.subplots(1, 2, figsize=(15,15))

ax[0].pie(g_1.groupby('host_is_superhost')['topic_5'].value_counts()['f'].sort_index(), 
                labels=[0,1,2,3,4], autopct='%1.2f%%', shadow=True)
ax[1].pie(g_1.groupby('host_is_superhost')['topic_5'].value_counts()['t'].sort_index(), 
                labels=[0,1,2,3,4], autopct='%1.2f%%', shadow=True)
ax[0].title.set_text('Superhost=False')
ax[1].title.set_text('Superhost=True')
plt.show()
```

```python
g_2 = pd.merge(item_listings, tmp, on='host_id')
```
